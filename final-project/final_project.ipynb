```python
import pandas as pd
import numpy as np

df = pd.read_csv(r"C:\Users\Isaac\Downloads\iris_extended.csv")
print(df)

print(df[df.isna().any(axis=1)]) #no NA values present

df.drop(['sepal_aspect_ratio', 'petal_aspect_ratio','sepal_petal_length_diff', 'sepal_petal_width_diff', 'sepal_area_sqrt', 'petal_area_sqrt'], axis=1, inplace=True)
# above columns dropped, they will be irrelevant to this analysis which is more concerned with the attractiveness of the flowers
```

```python
avg_sepal_area_by_species = df["sepal_area"].groupby(df['species']).agg("mean")
print(avg_sepal_area_by_species)

avg_petal_area_by_species = df["petal_area"].groupby(df['species']).agg("mean")
print(avg_petal_area_by_species)
# virginica species of iris demonstrates slightly higher average sepal area and much higher petal area. setosa has extremely small petal area - upon further research this seems consistent.

df.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sb


a = sb.barplot(data = df, x = 'species', y = 'petal_area')
plt.title('Average Petal Area by Species of Iris')
plt.ylabel('Petal Area')
plt.xlabel('Species')
for i in a.containers:
    a.bar_label(i,)
plt.legend()
#chart demonstrates that Iris virginica shows far greater petal area on average than setosa and versicolor
```

```python
sb.set_style('whitegrid')
a = sb.barplot(data = df, x = 'species', y = 'sepal_area')
plt.title('Average Sepal Area by Species of Iris')
plt.ylabel('Sepal Area')
plt.xlabel('Species')
for i in a.containers:
    a.bar_label(i,)
    


data_iv = df[df['species'] == 'virginica']
print(data_iv)
```

The above charts demonstrate that Iris virginica has higher sepal area on average than setosa and versicolor, and a much higher petal area.

Iris virginica will be used as species of choice. Petal and sepal area are a metric of focus because the sepal and petals are the showy assemblages of the flowering plant, and research shows
that larger petal and sepals are more attractive to both humans as well as pollinators, like native bees and butterflies. We could look at length and width separately, but looking at area accomplishes this and simplifies the work for us.

```python
avg_elev_virginica = df['elevation'].groupby(df['species'] == 'virginica').agg('mean')
print(avg_elev_virginica)
# avg elevation for virginica species = 174.1 m, non-virginica avg = 173.3 m
```

```python
sb.countplot(data = data_iv, x = df['soil_type'])
plt.ylabel('Count')
plt.xlabel('Soil Type')
plt.title('Frequency of Soil Type for Iris virginica')
# chart shows clay soil type to be most frequent, with loamy just behind and sandy just behin loamy.
```

```python
avg_petal_area_sandy_virginica = data_iv['petal_area'].groupby(data_iv['soil_type'] == 'sandy').agg('mean')
print(avg_petal_area_sandy_virginica)

# when in sandy soil, iris virginica has an average petal area of 9.47

avg_petal_area_clay_virginica = data_iv['petal_area'].groupby(data_iv['soil_type'] == 'clay').agg('mean')
print(avg_petal_area_clay_virginica)

# when in clay soil, iris virginica has an average petal area of 12.45

avg_petal_area_loamy_virginica = data_iv['petal_area'].groupby(data_iv['soil_type'] == 'loamy').agg('mean')
print(avg_petal_area_loamy_virginica)

# when in loamy soil, iris virginica has an average petal area of 10.85
```

Analysis of soil type shows clay soil to be most common, as well as producing a correlation with larger petal area than sandy or loamy soil. Because clay is naturally abundant, it is also usually a cheaper soil type, making this result good news for our florist.

```python
import sklearn
from sklearn.linear_model import LinearRegression
y = data_iv['sepal_area']
x = data_iv['elevation']

plt.scatter(x,y)
plt.xlabel('Elevation (m)')
plt.ylabel('Sepal Area (m^2)')
model = LinearRegression().fit(x.values.reshape(-1, 1), y)
r2 = model.score(x.values.reshape(-1, 1), y)
plt.plot(x, model.predict(x.values.reshape(-1, 1)), color='red')
plt.text(1, 4.5, f'R-squared = {r2:.2f}', fontsize=12)
```

```python
y = data_iv['petal_area']
x = data_iv['elevation']

plt.scatter(x,y)
plt.xlabel('Elevation (m)')
plt.ylabel('Petal Area (m^2)')
model = LinearRegression().fit(x.values.reshape(-1, 1), y)
r2 = model.score(x.values.reshape(-1, 1), y)
plt.plot(x, model.predict(x.values.reshape(-1, 1)), color='red')
plt.text(1, 4.5, f'R-squared = {r2:.2f}', fontsize=12)
```

Elevation was checked for any correlative effect on sepal and petal area, but there is no correlation indicated per R-square values of 0.03. It was checked because plant species who require higher elevation tend to need specific growth conditions, such as: cooler air, lower humidity, and some increased wind exposure/air circulation. Because the distribution suggests plants perform similarly across all elevations, we can feel more confident that the Iris virginica are easier for customers to grow due to a wide range of acceptable growth conditions - they will also be easier to grow in high volume for a florist.

```python
avg_sep_lengthratio = df["sepal_to_petal_length_ratio"].groupby(df['species']).agg("mean")
print(avg_sep_lengthratio)

avg_sep_widthratio = df["sepal_to_petal_width_ratio"].groupby(df['species']).agg("mean")
print(avg_sep_widthratio)
```

Sepal-to-petal length and width ratios are compared by species to get a greater idea of the appearance and shape of the flowers. Iris setosa has a very high ratio between the petals and sepals, meaning the sepals are far larger than the petals. Versicolor has a slightly larger sepal to petal length and sepals around twice the width as their petals. Iris virginica shows on average close to equally sized petals and sepals. Irises have three sepals which are usually large, and three petals which are usually small. Because Iris virginica has such large petals roughly the size of their showy sepals, they consequently look to have six large petals. This is a trait that could make them more charismatic and appear prettier, though subjective.
